```python
from crewai import Agent, Task, Crew , LLM
```

```python
llm=LLM(model="gpt-4o-mini")
```

# Building Agents Receipe Recommdeation

```python
main_ingerdient='tomato'
dietary_restriction='meat'
```

```python
ingerdient_assistant=Agent(llm=llm,
                           role="ingerdient_assistant",
                           backstory="An experienced ingerdient skilled in finding and tailoring reciepes based on ingerdient and dietary needs, and providing clear and step by step instruction ",
                           goal="Find receipes , filter them to meet dietary prefrences, and guide user through receipes steps",
                           verbose=True
                           )
```

# Assigning Task

```python
find_and_filter_receipes=Task(
    description=f"Find receipes that user the ingredient:{main_ingerdient} and "
    f"filter them to meet dietary restriction:{dietary_restriction}",
    expected_output=f"one receipes using {main_ingerdient} and matching {dietary_restriction} restriction.",
    agent=ingerdient_assistant
)

guide_reciepe_steps=Task(
    description="Provide step by step instruction for selected reciepe",
    expected_output=f"step by step instruction for the chosen recipe",
    agent=ingerdient_assistant
)
```

```python
crew=Crew(agents=[ingerdient_assistant],
          tasks=[find_and_filter_receipes, guide_reciepe_steps],
          planning=True)
```

```python
crew.kickoff()
```

```python
from crewai import Agent, Task, Crew , LLM
from crewai_tools import SerperDevTool
```

```python
product_name="energy_drink"
strategies_backstory=" and Marketing Strategy"
```

```python
from getpass import getpass
serper_api_key = getpass("Enter your Serper API key: ")
```

```python
import os
os.environ["SERPER_API_KEY"] = serper_api_key
```

```python
from getpass import getpass
openai_api_key = getpass("Enter your OpenAI API key: ")
```

```python
import os
os.environ["OPENAI_API_KEY"] = openai_api_key
```

```python
llm=LLM(model="gpt-4o-mini")
```

```python
market_researcher=Agent(llm=llm,
                        role="market_researcher",
                        goal="Analyze the market trends  for the product launch ",
                        backstory="Experienced in market a behaviour and customer behaviour analysis",
                        tools=[SerperDevTool(api_key=serper_api_key)],
                        verbose=True
                        )
```

```python
startgiest=Agent(llm=llm,
                        role="product_startgiest",
                        goal="Create effective positioning stratgies for the product ",
                        backstory=f"skilled in competitive positioning {strategies_backstory}",
                        verbose=True
                        )
```

```python
gather_market_insights_task=Task(
    description=f"Browse the internet to gather insights for the current market trends for the launch of {product_name}",
    expected_output=f"List of relevant market trends and consumer preferences, relevant to {product_name}",
    agent=market_researcher,
    verbose=True
)
```

```python
develop_position_stratgey_task=Task(
    description="Based on market insights,"f"create a positioning strategy for the {product_name} product , including analysis for impact and target audiance",
    expected_output=f"A positioning strategy with target audience and impact notes",
    agent=startgiest,
    verbose=True
)
```

```python
crew=Crew(agents=[market_researcher, startgiest],
          tasks=[gather_market_insights_task, develop_position_stratgey_task],
          planning=True)
```

```python
output=crew.kickoff()
```
